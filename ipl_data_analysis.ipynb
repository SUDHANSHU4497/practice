```python
# we have to find the top five batsman in terms of score
data3='/content/drive/MyDrive/dataset/deliveries.csv'
delivery=pd.read_csv(data3)
```

```python
# top 5 batsman who hit maximum number of 4s
mask=delivery['batsman_runs']==4
new_delivery=delivery[mask]
```

```python
new_delivery.shape[0]
```

```python
# total number of 6s
masks=delivery['batsman_runs']==6
new_deliveri=delivery[masks]
new_deliveri.shape[0]
```

```python
new_delivery.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head(5)
```

```python
# name the three team against which virat kohli scored maximum run
"""
step1 mask
step2 groupby team
step3 sum on batsman run"""

vk=delivery[delivery['batsman']=='V Kohli']
```

```python
vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(3)
```

```python
def run_scored(batsman_name):
    vk=delivery[delivery['batsman']==batsman_name]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(1).index[0]
```

```python
run_scored('V Kohli')
```
