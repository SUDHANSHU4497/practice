```python
def get_city_match_count(city):
    mask=data['city']==city # give boolean result
    return data[mask].shape[0]
```

```python
#value count method or function
# if in this dataset we have to find the number of  mathes win by every team
```

**value count method or function**
*   it works on series
*   work on categorical column
*   we can apply this method on numerical    column but it will not makes sense.




```python
# data.['winner']  give series
data['winner'].value_counts()
```

**plot function**


*   we can sketch multiple graph




bar chart


*   categorical data vs numerical data




```python
import matplotlib.pyplot as plt
# step1 fetch series data
data['winner']
#step2 now apply value_count method
data['winner'].value_counts()
#then type plot(kind='bar')
data['winner'].value_counts().plot(kind='bar')
```

```python
# for top 5 team
data['winner'].value_counts().head().plot(kind='bar')
```

```python
data['toss_decision'].value_counts().plot(kind='pie')
```


*   histogram work on numerical column
